```python
dm.data.reset_vector_cache()
```

```python
train, validation = dm.data.process(path='/home/common/aravind/deepmatch_100embedding/data/preprocess_data/2',
                                          train='train_dataset.csv',
                                          validation='val_dataset.csv',
                                          ignore_columns = ['left_id', 'right_id'],
                                          embeddings ='glove.6B.50d',
                                          embeddings_cache_path ='/mnt/.vector_cache')
```

```python
model = dm.MatchingModel(attr_summarizer='hybrid')
```

```python
model.run_train(
    train,
    validation,
    epochs=10,
    batch_size=16,
    best_save_path='/home/common/aravind/deepmatch_100embedding/data/model/paper_product_staples_walmart_e10_9960_50d.pth')
```

```python
model.save_state('/home/common/aravind/deepmatch_100embedding/data/model/paper_product_staples_walmart_e10_9960_50d_state.pth', include_meta=True)
```

```python
model.run_eval(validation)
```
